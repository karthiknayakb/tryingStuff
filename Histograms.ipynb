```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread("./Desert.jpg")
```

```python
color = ('b','g','r')
for i, col in enumerate(color):
    histogram = cv2.calcHist([image],[i],None,[256],[0,256])
    plt.plot(histogram,color=col)
    plt.xlim([0,256])
plt.show()
```
