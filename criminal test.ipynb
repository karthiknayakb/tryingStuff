```python
import pandas as pd
file = open('testfile.txt','r') 
cri_data = file.read().split(" \n")
```

```python
cri_test = pd.read_csv("criminal_test.csv")
columns = cri_test.columns
columns=columns[:-4].append(columns[-1:])
cri_test.head(5)
```

```python
columns
```

```python
def getstri(per):
    pp = ""
    for i in columns:
        if i=="PERID":
            continue
        pp+=str(float(per[i]))
    return pp
```

```python
result = []
getstri(cri_test.iloc[0])
testlen = len(cri_test)
for i in xrange(testlen):
    tem = (getstri(cri_test.iloc[i])).strip()
    if tem in cri_data:
        result.append([int(cri_test.iloc[i][columns[0]]),1])
    else:
        result.append([int(cri_test.iloc[i][columns[0]]),0])
```

```python
for i in xrange(10):
    tem = (getstri(cri_test.iloc[i])).strip()
    print tem," len:",len(tem)
    if tem in cri_data:
        result[int(cri_test.iloc[i][columns[0]])]=1
    else:
        result[int(cri_test.iloc[i][columns[0]])]=0
```

```python
col = ['PERID','Criminal']
resdf = pd.DataFrame(result)
resdf.columns=col
```

```python
resdf.to_csv("criminalRcsv.csv",index=False)
```
